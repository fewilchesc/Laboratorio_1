```python
from time import time
import numpy as np
import matplotlib.pyplot as plt
def fibonacci(n):
    instanteInicial = time()
    primero=0
    segundo=1
    resultado=0
    i = 0
    for i in range(n):
        resultado=primero+segundo
        segundo=primero
        primero=resultado
    instanteFinal = time()
    tiempoEjecucion=instanteFinal-instanteInicial
    return tiempoEjecucion


#modificar el valor de n
lista=[]
n=10000
for i in range(n):
    tiempo=fibonacci(i)
    lista.append(tiempo)
plt.plot(lista)
plt.xlabel('Tamano problema')
plt.ylabel('Tiempo Ejecucion')
```
